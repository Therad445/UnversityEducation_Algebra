```python
import matplotlib.pyplot as plt
import sympy as sp 
import numpy as np
from sympy.abc import x,y
```

## 1

```python

x1,y1=sp.symbols('x1 y1')
X1=sp.Matrix([[x1,y1]])
A = sp.Matrix([[1,-1],[-1,1]])
A1 = np.array([[1,-1],[-1,1]])
T,D=A.diagonalize()
L,T1=np.linalg.eig(A1)
print(T, T1, D, L, sep='\n')
X = T/sp.sqrt(2) * X1.T   # матрицу Т(sympy) переводим в нормированный вектор 
print('x=',sp.simplify(X[0,0]))
print('y=',sp.simplify(X[1,0]))



F = X[0]**2  -  2*X[0]*X[1]  + X[1]**2  -  10*X[0]  -  6*X[1]  +  25
F = sp.simplify(F)
F


sp.plot_implicit(sp.Eq(X[0]**2  -  2*X[0]*X[1]  + X[1]**2  -  10*X[0]  -  6*X[1]  +  25,0),title='В старой СК', xlabel='x',ylabel='y')
# sp.plot_implicit(sp.Eq(-8*sp.sqrt(2)*x + 2*y**2 + 2*sp.sqrt(2)*25 + 25 ),title='В новой СК', xlabel='x1',ylabel='y1')
sp.plot_implicit(sp.Eq(-8 * x+2*y**2 + 2*sp.sqrt(2)*y, 25),title='В новой СК', xlabel='x1',ylabel='y1')


```

```python
a = sp.Matrix([1,-1.5,0])
b = sp.Matrix([2,1,1])
c = sp.Matrix([3,0,4])
F = [a,b,c]
A = sp.Matrix([[1,-1.5,0], [2,1,1], [3,0,4]])
print(A.det())
b = sp.GramSchmidt(F)
print(b[0], b[1], b[2],sep = '\n')

fig = plt.figure()
ax = plt.axes(projection ='3d')

plt.quiver(0,0,0,1,-1.5,0)
plt.quiver(0,0,0,2,1,1)
plt.quiver(0,0,0,3,0,4)

print(b[1].dot(b[0]))
print(b[2].dot(b[0]))
print(b[2].dot(b[1]))
```
