```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols
from sympy import *

x1,y1,alfa = symbols('x1 y1 alfa')
x = x1 * cos(alfa) - y1 * sin(alfa)
y = y1 * cos(alfa) + x1 * sin(alfa)
f = y - 2 * x - 4
# решаем систему уравнений
y1 = sy.solve(f,'y1')
print(y1)
y1 = x1 * (-x1*sin(alfa) + 2*x1*cos(alfa) + 4)/(2*sin(alfa) + cos(alfa))
yy = y1.subs(alfa,pi / 4)
print('Уравнение повернутой прямой:', yy)

alfa = -np.pi/4
a = 3
b = 2

x0 = np.linspace(-1, 6, 100)
y0 = 2*x0 - 4

plt.plot(x0, y0, '-r')

x = x0 * np.cos(alfa) - y0 * np.sin(alfa) + a
y = y0 * np.cos(alfa) + x0 * np.sin(alfa) + b

plt.plot(x, y, '--g')

plt.arrow(-1, 0, 6, 0, width=0.08, color='black')
plt.arrow(0, -2, 0, 7, width=0.08, color='black')
x3 = np.cos(-alfa); y3 = - np.sin(-alfa)
x4 = np.sin(-alfa); y4 = np.cos(-alfa)
plt.arrow(a, b, x3, y3, width=0.08, color='blue')
plt.arrow(a, b, x4, y4, width=0.08, color='blue')

x1 = (1 - a) * np.cos(alfa) + (-2 - b) * np.sin(alfa)
y1 = (-2 - b) * np.cos(alfa) - (1 - a) * np.sin(alfa)
x11 = (1) * np.cos(alfa) - (-2) * np.sin(alfa) + a
y11 = (-2) * np.cos(alfa) + (1) * np.sin(alfa) + b

print(x1, y1,'координаты точки в новой синей системе')

plt.plot(1, -2, 'ob')
plt.text(1 + 0.2, -2, 'M0')
plt.plot(x11, y11, 'og')
plt.text(x11 + 0.2, y11, 'M1')

plt.xlabel('x'); plt.ylabel('y')
plt.axis('equal')
plt.grid()
plt.show()
```
