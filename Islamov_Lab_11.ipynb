```python
import numpy as np
from sympy import *
#  
A = np.array([[1, 2, 3], [1, -3, 2], [2, 4, 6]])
B = np.array([7,5,14])
#   
D = np.column_stack((A,B)) #      
n=3
rkA=np.linalg.matrix_rank(A)
rkD=np.linalg.matrix_rank(D)
print(' :',[rkA,rkD,n])

```

```python
#  
A = np.array([[3, -2, 1,1], [5, 1, 2,0], [-1,1,-1,1],[2,-1,6,-3]])
B = np.array([-8,-11,0,9])
#   
D = np.column_stack((A,B)) #      
n=4
rkA=np.linalg.matrix_rank(A)
rkD=np.linalg.matrix_rank(D)
print(' :',[rkA,rkD,n])
X1=np.linalg.inv(A).dot(B)
print(':',X1)
#  
DA=np.linalg.det(A)
D1=A.copy()
D1[:,0]=B
D11=np.linalg.det(D1)
D2=A.copy()
D2[:,1]=B
D22=np.linalg.det(D2)
D3=A.copy()
D4=A.copy()
D3[:,2]=B
D33=np.linalg.det(D3)
D4[:,3]=B
D44=np.linalg.det(D4)
X2=[D11/DA,D22/DA,D33/DA,D44/DA]
print(':',X2) #    - ,  ,      0.
#  solve
X3 = np.linalg.solve(A,B)
print(':',X3)
X3[0]=-3
X3[1]=0
X3[2]=2
X3[3]=-1
# 
np.dot(A,X3)==B
```
