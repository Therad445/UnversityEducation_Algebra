```python
import numpy as np
import matplotlib.pyplot as plt
import math	
from sympy import *
```

Упражнение 3.1

```python
OA = np.array([-2, 0])
OB = np.array([1, 2])
OC = np.array([1, -1])
OD = np.array([-2,-3])
print('Векторы имеют координаты:')
AB = OB - OA
BC = OC - OB
CD = OD - OC
AC = OC - OA
AD = OD - OA
print('AB:', AB)
print('BC:', BC)
print('CD:', CD)
print('AC:', AC)
print('AD:', AD)

plt.grid()
plt.ylim(-4, 3)
plt.xlim(-3, 2)
plt.arrow(-2, 0, AB[0], AB[1], width=0.07, length_includes_head=True, facecolor='Yellow')
plt.arrow(1, 2, BC[0], BC[1], width=0.07, length_includes_head=True, facecolor='Red')
plt.arrow(-2, 0, AC[0], AC[1], width=0.07, length_includes_head=True, facecolor='Blue')
plt.arrow(-2, 0, AD[0], AD[1], width=0.07, length_includes_head=True, facecolor='Green')
plt.arrow(1, -1, CD[0], CD[1], width=0.07, length_includes_head=True, facecolor='Black')
plt.show()
```

#Упражнение 3.4

```python
A = np.array([-2, 0])
B = np.array([3, 4])
C = np.array([4, -1])

cos_AB = np.dot(A, B) / np.linalg.norm(A) / np.linalg.norm(B)
print('cos(A,B)', cos_AB)
ugol1= np.arccos(cos_AB) # арккосинус угла
ugol1=math.degrees(ugol1)
print('Угол в градусах', ugol1)

cos_BC = np.dot(B, C) / np.linalg.norm(B) / np.linalg.norm(C)
print('cos(B,C)', cos_BC)
ugol2= np.arccos(cos_BC) # арккосинус угла
ugol2=math.degrees(ugol2)
print('Угол в градусах', ugol2)

cos_CA = np.dot(C, A) / np.linalg.norm(C) / np.linalg.norm(A)
print('cos(C,A)', cos_CA)
ugol3= np.arccos(cos_CA) # арккосинус угла
ugol3=math.degrees(ugol3)
print('Угол в градусах', ugol3)


OA = np.array([-2, 0])
OB = np.array([3, 4])
OC = np.array([4, -1])

print('Векторы имеют координаты:')

AB = OB - OA
BC = OC - OB
AC = OC - OA
print('AB:', AB)
print('BC:', BC)
print('AC:', AC)

plt.grid()
plt.ylim(-5, 5)
plt.xlim(-3, 5)
plt.arrow(-2, 0, AB[0], AB[1], width=0.07, length_includes_head=True, facecolor='Yellow')
plt.arrow(3, 4, BC[0], BC[1], width=0.07, length_includes_head=True, facecolor='Red')
plt.arrow(4, -1, -AC[0], -AC[1], width=0.07, length_includes_head=True, facecolor='Blue')
plt.show()
```
