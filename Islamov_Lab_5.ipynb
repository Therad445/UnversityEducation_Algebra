```python
a=Plane((1,-1,-3),(2,-3,5))
print(a.equation())
A=2;B=-3;C=5;D=10
f = lambda x, y: (-A * x -B * y -  D) / C 
fig = plt.figure(figsize=(10, 10)) 
ax = fig.add_subplot(1, 1, 1, projection='3d') #
xval = np.linspace(-4, 4, 100) 
yval = np.linspace(-4, 4, 100) 
x, y = np.meshgrid(xval, yval)
p = np.array([1, -1, - 3]) 
v = np.array([2, -3, 5]) 
ax.quiver(*p, *v, color='Black') 
z = f(x, y)
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10, cmap=cm.bone_r) 
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плоскость 2x - 3y + 5z + 10 = 0')
ax.view_init(10, -30)
plt.show()
```

```python
M = (2,-1,-3)
A = (-1,2,0)
B =(2,-4,0) #B = sp.Point(1, -2, 1)
L1 = Line(A,B)
M1 = L1.parallel_line(M)
M1 = M1.arbitrary_point()
print('Параметрическое уравнение прямой', M1)



#плоскость
A = 5; B = 3; C = -9; D = -55
f = lambda x, y: (-A * x - B * y - D) / C # строим плоскость
fig = plt.figure(figsize = (10, 10)) # создаем холст
ax = fig.add_subplot(1, 1, 1, projection = '3d')
xval = np.linspace(-8, 8, 100) # разбиение для х
yval = np.linspace(-8, 8, 100) # разбиение для y 
x, y = np.meshgrid(xval, yval)
ax.scatter(2, 3, -4, s = 120.0) #точка, из которой выходит нормальный вектор
M = np.array([2, 3, -4]) # точка на плоскости
n = np.array([5, 3, -9]) # вектор нормали
ax.quiver(*M, *n, color = 'Blue') # строим вектор нормали
z = f(x, y) # вычисляем значения z в точках
surf = ax.plot_surface(x, y, z, rstride = 10, cstride = 10, cmap = cm.plasma).set_alpha(0.3) # строим плоскость
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плоскость 5x + 3y - 9z - 55 = 0')
ax.view_init(10, -30)
plt.show()
```
