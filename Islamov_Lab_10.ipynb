```python
# Построение эллипсоида в сферических координатах
fig = plt.figure(figsize=(7, 7)) # создаём холст
ax = fig.add_subplot(111, projection='3d')
coefs = (6, 3, 2)# коэффициенты уравнения
rx, ry, rz = 1/np.sqrt(coefs) # радиусы
# Сферические углы:
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0, np.pi, 100)
# Уравнение эллипсоида:
x = rx*np.outer(np.cos(u), np.sin(v))
y = ry*np.outer(np.sin(u), np.sin(v))
z = rz*np.outer(np.ones_like(u), np.cos(v))
ax.plot_surface(x,y,z, rstride=5, cstride=5, color='m')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Эллипсоид')
plt.show()
```
